```python
import pandas as pd
from statsbombpy import sb
from reportlab.lib.pagesizes import letter
from reportlab.lib import colors
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle, KeepTogether
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle

pdf_filename = "statsbomb_tournament_report.pdf"
doc = SimpleDocTemplate(
    pdf_filename,
    pagesize=letter,
    rightMargin=40, leftMargin=40, topMargin=40, bottomMargin=40
)

styles = getSampleStyleSheet()

title_style = ParagraphStyle(
    'DocTitle', parent=styles['Heading1'],
    fontSize=22, leading=26, textColor=colors.HexColor("#1a365d"), spaceAfter=4
)
subtitle_style = ParagraphStyle(
    'DocSubtitle', parent=styles['Normal'],
    fontSize=10, leading=14, textColor=colors.HexColor("#4a5568"), spaceAfter=15
)
match_header_style = ParagraphStyle(
    'MatchHeader', parent=styles['Heading2'],
    fontSize=13, leading=16, textColor=colors.HexColor("#2d3748"), spaceBefore=10, spaceAfter=2
)
match_id_style = ParagraphStyle(
    'MatchID', parent=styles['Normal'],
    fontSize=8, leading=10, textColor=colors.HexColor("#a0aec0"), spaceAfter=8
)
section_title_style = ParagraphStyle(
    'SectionTitle', parent=styles['Heading3'],
    fontSize=9, leading=12, textColor=colors.HexColor("#2b6cb0"), spaceBefore=6, spaceAfter=4
)
body_style = ParagraphStyle(
    'TableBody', parent=styles['Normal'],
    fontSize=8.5, leading=11, textColor=colors.HexColor("#2c3e50")
)
header_cell_style = ParagraphStyle(
    'TableHeaderCell', parent=styles['Normal'],
    fontSize=8, leading=10, textColor=colors.HexColor("#4a5568"), fontName="Helvetica-Bold"
)
score_style = ParagraphStyle(
    'ScoreStyle', parent=styles['Normal'],
    fontSize=10, leading=12, textColor=colors.white, fontName="Helvetica-Bold", alignment=1
)
empty_style = ParagraphStyle(
    'EmptyStyle', parent=styles['Normal'],
    fontSize=8.5, leading=11, textColor=colors.HexColor("#a0aec0"), fontName="Helvetica-Oblique"
)

table_header_style = TableStyle([
    ('BACKGROUND', (0, 0), (-1, 0), colors.HexColor("#f7fafc")),
    ('BOTTOMPADDING', (0, 0), (-1, -1), 4),
    ('TOPPADDING', (0, 0), (-1, -1), 4),
    ('ALIGN', (0, 0), (-1, -1), 'LEFT'),
    ('LINEBELOW', (0, 0), (-1, -1), 0.5, colors.HexColor("#edf2f7")),
])

story = []

story.append(Paragraph("Tournament Match Event Log", title_style))
story.append(Paragraph("Parsed StatsBomb Match Statistics • Goals, Assists, and Live Summaries", subtitle_style))
story.append(Spacer(1, 10))

for index, match_id in enumerate(match_ids_array, start=1):
    match_elements = []
    
    events = sb.events(match_id=match_id)
    
    teams = list(events['team'].dropna().unique())
    if len(teams) >= 2:
        team_a, team_b = teams[0], teams[1]
        team_names = f"{team_a} vs {team_b}"
    else:
        team_a, team_b = "Unknown Team 1", "Unknown Team 2"
        team_names = "Unknown Teams"
        
    match_elements.append(Paragraph(f"Match {index}: {team_names}", match_header_style))
    match_elements.append(Paragraph(f"Match ID: {match_id}", match_id_style))
    
    match_elements.append(Paragraph("⚽ Goals & Own Goals", section_title_style))
    
    # Separate normal game goals from penalty shootout goals (Period 5)
    goals = events[(events['shot_outcome'] == 'Goal') & (events['period'] < 5)]
    own_goals = events[events['type'].isin(['Own Goal Against'])]
    
    if len(goals) == 0 and len(own_goals) == 0:
        match_elements.append(Paragraph("No goals scored in regulation/extra time.", empty_style))
    else:
        data = [[Paragraph("Min", header_cell_style), Paragraph("Sec", header_cell_style), 
                 Paragraph("Player", header_cell_style), Paragraph("Team", header_cell_style), 
                 Paragraph("xG / Type", header_cell_style)]]
                 
        for _, r in goals.iterrows():
            shot_type_label = f" ({r['shot_type']})" if 'shot_type' in r and r['shot_type'] == 'Penalty' else ""
            xg = f"{r['shot_statsbomb_xg']:.2f}{shot_type_label}" if 'shot_statsbomb_xg' in r and not pd.isna(r['shot_statsbomb_xg']) else f"-{shot_type_label}"
            data.append([Paragraph(f"{int(r['minute'])}'", body_style), Paragraph(f"{int(r['second'])}\"", body_style),
                         Paragraph(str(r['player']), body_style), Paragraph(str(r['team']), body_style), Paragraph(xg, body_style)])
                         
        for _, r in own_goals.iterrows():
            data.append([Paragraph(f"{int(r['minute'])}'", body_style), Paragraph(f"{int(r['second'])}\"", body_style),
                         Paragraph(str(r['player']), body_style), Paragraph(str(r['team']), body_style), Paragraph("OG", body_style)])
                         
        t = Table(data, colWidths=[35, 35, 200, 160, 100])
        t.setStyle(table_header_style)
        match_elements.append(t)
        
    match_elements.append(Spacer(1, 6))
        
    match_elements.append(Paragraph("🎯 Assists", section_title_style))
    if 'pass_goal_assist' in events.columns:
        assists = events[(events['type'] == 'Pass') & (events['pass_goal_assist'] == True) & (events['period'] < 5)]
    else:
        assists = []
        
    if len(assists) == 0:
        match_elements.append(Paragraph("No assists recorded.", empty_style))
    else:
        data = [[Paragraph("Min", header_cell_style), Paragraph("Sec", header_cell_style), 
                 Paragraph("Player", header_cell_style), Paragraph("Team", header_cell_style), 
                 Paragraph("Recipient", header_cell_style)]]
                 
        for _, r in assists.iterrows():
            recipient = r['pass_recipient'] if 'pass_recipient' in r else '-'
            data.append([Paragraph(f"{int(r['minute'])}'", body_style), Paragraph(f"{int(r['second'])}\"", body_style),
                         Paragraph(str(r['player']), body_style), Paragraph(str(r['team']), body_style), Paragraph(str(recipient), body_style)])
                         
        t = Table(data, colWidths=[35, 35, 200, 130, 130])
        t.setStyle(table_header_style)
        match_elements.append(t)
        
    # Calculate scores for regulation + extra time
    score_a = len(goals[goals['team'] == team_a]) + len(own_goals[own_goals['team'] == team_b])
    score_b = len(goals[goals['team'] == team_b]) + len(own_goals[own_goals['team'] == team_a])
    
    score_text = f"FINAL SCORE: {team_a} {score_a} - {score_b} {team_b}"
    
    # Check if a penalty shootout occurred (Period 5)
    if 'period' in events.columns and 5 in events['period'].values:
        shootout_goals = events[(events['shot_outcome'] == 'Goal') & (events['period'] == 5)]
        pens_a = len(shootout_goals[shootout_goals['team'] == team_a])
        pens_b = len(shootout_goals[shootout_goals['team'] == team_b])
        score_text += f" (Pens: {pens_a} - {pens_b})"
    
    match_elements.append(Spacer(1, 8))
    score_table = Table([[Paragraph(score_text, score_style)]], colWidths=[530])
    score_table.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, -1), colors.HexColor("#2b6cb0")),
        ('TOPPADDING', (0, 0), (-1, -1), 6),
        ('BOTTOMPADDING', (0, 0), (-1, -1), 6),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
    ]))
    match_elements.append(score_table)
    match_elements.append(Spacer(1, 20))
    
    story.append(KeepTogether(match_elements))

doc.build(story)
print(f"Successfully generated PDF report layout via ReportLab saved as: '{pdf_filename}'")
```
